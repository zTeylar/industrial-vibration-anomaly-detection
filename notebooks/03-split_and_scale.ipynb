```python
import numpy as np
from pathlib import Path


training_set = np.load('../datasets/training_set.npy')[: ,: ,: 32]
test_set = np.load('../datasets/testing_set.npy')[: ,: ,: 32]

#CNNs require a 4D array that includes the color, we will be using a 1 to indicate our Greyscale
training_set = np.expand_dims(training_set, axis=-1)
test_set = np.expand_dims(test_set, axis=-1)

concatenated = np.concatenate((training_set, test_set), axis=0)
training_set = concatenated[:13583]
test_set = concatenated[13583:]


train_min = training_set.min()
train_max = training_set.max()

def min_max(data, global_min, global_max):
    norm_dataset = (data - global_min) / (global_max - global_min)
    return norm_dataset

normalized_training = min_max(training_set, train_min, train_max)
normalized_testing = min_max(test_set,train_min, train_max)

# print(normalized_training.min(), normalized_training.max(), normalized_testing.min(), normalized_testing.max())
# print(train_min, train_max, test_set.min(), test_set.max())
print(normalized_training.shape)
```

```python
output_dir = Path('../datasets/normalized')
output_dir.mkdir(parents=True, exist_ok=True)
np.save(output_dir/'norm_training_set.npy', normalized_training)
np.save(output_dir/'norm_testing_set.npy', normalized_testing)
np.save(output_dir/'train_min.npy', train_min)
np.save(output_dir/'train_max.npy', train_max)
```
